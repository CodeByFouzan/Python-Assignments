```python
text = input("Enter a string: ")
uppercase = 0
lowercase = 0
digits = 0 
spaces = 0
for char in text:
    if char.isupper():
        uppercase += 1
    elif char.islower():
        lowercase += 1
    elif char.isdigit():
        digits += 1
    elif char.isspace():
        spaces += 1
print("Number of uppercase letters:", uppercase)
print("Number of lowercase letters:", lowercase)
print("Number of digits:", digits)
print("Number of whitespace characters:", spaces)
```

```python
text = input("Enter a string: ")
firstletter = text[0]
secondletter = text[len(text)-1]
newstring = secondletter + text[1:len(text) - 1] + firstletter
print(newstring)
```

```python
text = input("Enter a string: ")
reversed_text = ""
for i in range(len(text) - 1, -1, -1):
    reversed_text += text[i]
print("Reversed string:", reversed_text)
```

```python
name = input("Enter your name: ")
initials = name[0].upper() + ". "
for i in range(len(name)):
    if name[i] == ' ':
        initials += name[i+1].upper() + ". "
print("Your initials are:", initials)
```

```python
word = input("Enter a word: ").lower()
if word == word[::-1]:
    print("It's a palindrome!")
else:
    print("It's NOT a palindrome!")
```

```python
password = input("Enter your password: ")
is_long_enough = len(password) >= 8
has_uppercase = False
has_lowercase = False
has_digit = False
for char in password:
    if char.isupper():
        has_uppercase = True
    elif char.islower():
        has_lowercase = True
    elif char.isdigit():
        has_digit = True
if is_long_enough and has_uppercase and has_lowercase and has_digit:
    print("Password is valid!")
else:
    print("Password is invalid. Please make sure:")
    if not is_long_enough:
        print("- Password must be at least 8 characters long")
    if not has_uppercase:
        print("- Password must contain at least one uppercase letter (A-Z)")
    if not has_lowercase:
        print("- Password must contain at least one lowercase letter (a-z)")
    if not has_digit:
        print("- Password must contain at least one numeric digit (0-9)")
```
