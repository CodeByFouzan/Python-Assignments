```python
# Qno.3
user_input = input("Enter elements of the list separated by spaces: ")
my_list = [int(x) for x in user_input.split()]  # List comprehension concept

Max = 0
for num in my_list:
    if num > Max:
        Max = num

print("The highest number is:", Max)
        
```

```python
#Q4
List_Input = input("Enter a list of numbers in CSV format: ")
List_Split = [int(x) for x in List_Input.split(",")]
print("Original:", List_Split)

# Make a copy of the original list so to not again and agan copy the same value on every index
original = List_Split.copy()

for x in range(len(List_Split)):
    if x == 0:
       List_Split[x] = original[-1]
    else:
        List_Split[x] = original[x-1]

print("After:", List_Split)
```

```python
#Q9
def add_matrices(A, B):
    rows = len(A)
    cols = len(A[0]) 
    if rows != len(B) or cols != len(B[0]):
        print("Matrices are not of the same dimensions")
    result = []
    for i in range(rows):
        row = []
        for j in range(cols):
            row.append(A[i][j] + B[i][j])
        result.append(row)
    
    return result

A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8, 9],
     [10, 11, 12]]

result = add_matrices(A, B)
print("Matrix Addition Result:")
for row in result:
    print(row)
```

```python
#Q10
def multiply_matrices(A, B):
    n = len(A)
    m = len(A[0])
    p = len(B[0])

    # Check if multiplication is possible
    if m != len(B):
        return "Number of columns in A must be equal to number of rows in B"

    # Create result matrix with size n x p
    result = []
    for i in range(n):
        row = []
        for j in range(p):
            sum_product = 0
            for k in range(m):
                sum_product += A[i][k] * B[k][j]
            row.append(sum_product)
        result.append(row)

    return result

# Example
A = [[1, 2],
     [3, 4],
     [5, 6]]

B = [[7, 8, 9],
     [10, 11, 12]]

result = multiply_matrices(A, B)
print("Matrix Multiplication Result:")
for row in result:
    print(row)
```
